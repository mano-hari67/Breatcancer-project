**1. Explain missing values handling.**

1.Missing Values Handling

Missing values in Pandas are usually represented as NaN (Not a Number).

Common methods:

isnull() / notnull() – Detect missing values

dropna() – Remove rows/columns with missing values

fillna() – Replace missing values

Example:

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    "A": [1, 2, None, 4],
    "B": [5, None, 7, 8]
})

# Check missing values
df.isnull()

# Drop rows with missing values
df.dropna()

# Fill missing values with 0
df.fillna(0)

# Fill missing values with column mean
df.fillna(df.mean())

```

**2. Explain groupby.**

groupby() is used to group data based on one or more columns and apply aggregate functions.

Common aggregations:

sum(), mean(), count(), min(), max()

Example:

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Department": ["HR", "IT", "IT", "HR", "Finance"],
    "Salary": [30000, 50000, 55000, 32000, 60000]
})

# Group by Department and calculate mean salary
df.groupby("Department")["Salary"].mean()

```

**3. Explain merge and join.**

Definition

merge() combines DataFrames using one or more common columns as keys.

Syntax

pd.merge(left_df, right_df, on="column_name", how="type")


#Types of Merge

1.Inner Merge

Returns only rows with matching keys in both DataFrames.

```python
import pandas as pd

# Create first DataFrame
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["A", "B", "C"]
})

# Create second DataFrame
df2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Salary": [30000, 40000, 50000]
})

# Inner merge on ID column
result = pd.merge(df1, df2, on="ID", how="inner")
print(result)

```

2.Left Merge

Returns all rows from the left DataFrame and matching rows from the right.

```python
pd.merge(df1, df2, on="ID", how="left")
```

3.Right Merge

Returns all rows from the right DataFrame.

```python
pd.merge(df1, df2, on="ID", how="right")
```

4.Outer Merge

Returns all rows from both DataFram

```python
pd.merge(df1, df2, on="ID", how="outer")
```

**5. Explain sorting.**

Sorting helps arrange data in a specific order.

Types:

sort_values() – Sort by column

sort_index() – Sort by index

Example:

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Name": ["A", "B", "C", "D"],
    "Marks": [75, 88, 62, 90]
})

# Sort by Marks in descending order
df_sorted_values = df.sort_values(by="Marks", ascending=False)

# Sort by index
df_sorted_index = df.sort_index()

print(df_sorted_values)
print(df_sorted_index)

```

**6. Explain filtering.**

Filtering extracts specific rows based on conditions.

Example:

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Name": ["A", "B", "C", "D"],
    "Age": [17, 19, 21, 18],
    "Marks": [65, 78, 85, 92],
    "Grade": ["B", "A", "A", "A"]
})

# Filter students with Age > 18
result_age = df[df["Age"] > 18]

# Filter students with Marks > 70 and Grade = 'A'
result_marks_grade = df[(df["Marks"] > 70) & (df["Grade"] == "A")]

print(result_age)
print(result_marks_grade)

```

**7. Explain apply function.**

apply() applies a custom function along rows or columns.

Example:

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Salary": [30000, 40000, 50000]
})

# Apply function to calculate Bonus
df["Bonus"] = df["Salary"].apply(lambda x: x * 0.10)

print(df)

```

**8. Explain pivot table.**

8.Pivot Table

Pivot tables summarize data using aggregation.

Example:

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Region": ["North", "North", "South", "South"],
    "Year": [2023, 2024, 2023, 2024],
    "Sales": [1000, 1500, 1200, 1800]
})

# Create pivot table
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Region",
    columns="Year",
    aggfunc="sum"
)

print(pivot)

```

**9. Explain time series.**

9.Time Series

Pandas supports time-based data using datetime.

Common operations:

Date parsing

Resampling

Time shifting

Example:

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Date": ["2024-01-10", "2024-01-20", "2024-02-15", "2024-02-25"],
    "Sales": [100, 200, 150, 300]
})

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Set Date as index
df.set_index("Date", inplace=True)

# Resample monthly and calculate mean
monthly_mean = df.resample("M").mean()

print(monthly_mean)

```

**10. Explain data consistency.**

#Definition

Data consistency means ensuring that all data values are uniform, accurate, and follow the same format across the dataset.
Consistent data ensures reliability, accuracy, and integrity in analysis, reporting, and machine learning models.

#How to Ensure Data Consistency

#1.Standardize formats

 >Convert all dates to a single format: YYYY-MM-DD

 >Convert text to lower or upper case

#2.Use consistent units

 >All weights in kg, all prices in USD

#3.Validate data

 Use constraints, allowed values, or regular expressions

#4.Check across tables

 >Ensure foreign keys and references match

#5.Automate cleaning

 >Use Pandas functions, scripts, or database constraints

```python
#Pandas Examples for Consistency
import pandas as pd

# Sample inconsistent data
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Gender': ['F', 'Male', 'M'],
    'DOB': ['12-05-1990', '1991/07/23', '15-08-1989'],
    'Marks': [85, 105, 95]  # Notice 105 is invalid
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# -------------------------------
# 1. Standardize Gender column
df['Gender'] = df['Gender'].replace({'F': 'Female', 'M': 'Male'})
# Convert all text to proper case if needed
df['Gender'] = df['Gender'].str.capitalize()

# -------------------------------
# 2. Validate Marks (should be between 0 and 100)
df = df[df['Marks'].between(0, 100)]

# -------------------------------
# 3. Standardize DOB to YYYY-MM-DD format
df['DOB'] = pd.to_datetime(df['DOB'], dayfirst=True).dt.strftime('%Y-%m-%d')

print("\nCleaned and Consistent Data:")
print(df)


```

**1.What is feature engineering?**

- Definition: The process of creating, transforming, or selecting variables (features) to improve model performance.
- Goal: Make raw data more useful for algorithms.
- Example: From a “date” column, engineer features like day of week, month, or holiday flag


**2.Explain feature scaling.**

- Purpose: Standardize the range of independent variables so no feature dominates others.
- Techniques:
- Normalization: Scale values between 0 and 1.
- Standardization: Transform data to have mean = 0 and standard deviation = 1.
- Example: Scaling height (cm) and weight (kg) so both contribute equally in distance-based models like KNN.


**3.Explain feature extraction.**

- Definition: Creating new features from raw data, often reducing dimensionality.
- Example: Extracting edges or textures from images, or frequency components from audio.
- Tools: PCA, Fourier Transform, CNN filters


**4.Explain feature selection.**

- Purpose: Choosing the most relevant features and discarding irrelevant/redundant ones.
- Methods:
- Filter (correlation, chi-square)
- Wrapper (recursive feature elimination)
- Embedded (Lasso, decision trees)
- Benefit: Reduces overfitting, improves efficiency


**5.Explain encoding techniques.**

- Purpose: Convert categorical data into numerical form.
- Types:
- One-Hot Encoding: Creates binary columns for each category.
- Label Encoding: Assigns integer values to categories.
- Target Encoding: Replace category with mean target value.
- Example: “Color = Red, Blue, Green” → One-hot encoded as [1,0,0], [0,1,0], [0,0,1].


**6.Explain binning.**

- Definition: Grouping continuous values into discrete intervals.
- Example: Age → bins like 0–18, 19–35, 36–60, 60+.
- Use: Reduces noise, handles outliers, simplifies models


**7.Explain log transformation.**

- Purpose: Reduce skewness and handle wide-ranging values.
- Example: Income data often has extreme values; applying log(x) makes distribution closer to normal.
- Benefit: Improves performance of linear models


**8.Explain interaction features.**

- Definition: Features created by combining two or more existing features.
- Example: In housing data, Rooms × Area might be more predictive than each alone.
- Use: Captures relationships between variables.


**9.Explain dimensionality reduction.**

- Purpose: Reduce number of features while retaining most information.
- Techniques:
- PCA (Principal Component Analysis)
- t-SNE, UMAP for visualization
- Benefit: Faster training, less overfitting, easier visualization


**10.Explain importance of features.**

- Definition: Measures how much each feature contributes to predictions.
- Methods:
- Tree-based models (feature importance scores)
- Permutation importance
- SHAP values (explainable AI)
- Example: In credit scoring, income and payment history may be more important than age.
